### Файл 11.  Нарисовать отдельнтные импульсы с указанием событий и каналов.

```python
i = 11
filename = name + str(i) + ".txt"
datum = pd.read_csv(filename, header=None, index_col=False, sep='\s+')

n = "193 4539 8843 14931 19094 19825 25244 27396 30727"
n = list(map(int, n.split()))

fig, axn = plt.subplots(len(n), 1, figsize=(10, 15))
axn[0].set_title("file  " + filename[8:-4])

iplot = 0
for num in range(0, datum.shape[0], 1):
    data = datum.iloc[num] - datum.iloc[num].mean()
    if data.max() - data.min() > 10:
        axn[iplot].plot(data, label="Event " + str(num // 12) + " - chan " + str(num % 12))
        axn[iplot].set_xlim(0, len(datum.iloc[num]))
        axn[iplot].legend()
        axn[iplot].grid()
        iplot += 1
plt.savefig("file11.pulses.png", bbox_inches='tight')
```
